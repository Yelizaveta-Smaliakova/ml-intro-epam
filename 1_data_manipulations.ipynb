**Task 8 (1.25 point).** Find a player who participated in the largest number of matches during the whole match history. Assign a `player_name` to the given variable

```python
matches_players = match_data.loc[:,'home_player_1':'away_player_11'].to_numpy().flatten().tolist()
matches_players = [x for x in matches_players if math.isnan(x) == False]
matches_players = [int(x) for x in matches_players]
matches_players.sort()

matches_players_and_games = [[matches_players[0],1]]

for i in range (1,len(matches_players)):
    if matches_players_and_games[-1][0] == matches_players[i]:
        matches_players_and_games[-1][1] += 1
    else: 
        matches_players_and_games.append([matches_players[i],1])

matches_players_and_games.sort(key=lambda x: x[1], reverse=True)

max_matches_player = player_data[player_data['player_api_id'] == matches_players_and_games[0][0]].iloc[0]['player_name']
max_matches = matches_players_and_games[0][1]

for i in range (1,len(matches_players_and_games)):
    player = player_data[player_data['player_api_id'] == matches_players_and_games[i][0]].iloc[0]['player_name']
    matches = matches_players_and_games[i][1]
    if matches == max_matches:
        if max_matches_player > player:
            max_matches_player = player
    else:
        break

print(max_matches_player)
```

**Task 10 (2 points).** Find top-5 most similar players to **Neymar** whose names are given. The similarity is measured as [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) between vectors of players' characteristics (described in the task above). Put their names in a vector in ascending order by Euclidean distance and sorted by `player_name` if the distance is the same<br>
**Note 1:** There are many records for some players in the `Player_Attributes` table. You need to take the freshest data (characteristics with the most recent `date`). <br>
**Note 2:** Use pure values of the characteristics even if you are aware of such preprocessing technics as normalization. <br>
**Note 3:** Please avoid using any built-in methods for calculating the Euclidean distance between vectors, think about implementing your own.

```python
def euclidian_distance(first_player, second_player):
    delta = 0
    for i in range (len(first_player)):
        delta += pow(first_player[i] - second_player[i], 2)
    return pow(delta, 1/2)
```

```python
def player_info(id):
    attributes = player_attributes[player_attributes['player_api_id'] == id].sort_values(by='date', ascending=False).head(1)
    player_vector = attributes.loc[:,'crossing':'gk_reflexes'].to_numpy()
    player_vector = np.append(player_vector,attributes.loc[:,'overall_rating':'potential'].to_numpy())
    return player_vector
```

```python
neymar_id = player_data[player_data['player_name']=='Neymar'].iloc[0]['player_api_id']
neymar_vector = player_info(neymar_id)
```

```python
player_euclidian_distance_from_neymar = []

for i in range (len(player_data)):
    id = player_data.iloc[i]['player_api_id']
    vector = player_info(id)
    eu_dis = euclidian_distance(neymar_vector, vector)
    if math.isnan(eu_dis) == False:
        player_euclidian_distance_from_neymar.append([player_data.iloc[i]['player_name'],eu_dis])
```

```python
player_euclidian_distance_from_neymar.sort(key=lambda x: x[1], reverse=False)
```

```python
neymar_similarities = []
for i in range (1,6):
    neymar_similarities.append(player_euclidian_distance_from_neymar[i][0])

print(neymar_similarities)
```

```python
assert(len(neymar_similarities) == 5)
assert(isinstance(neymar_similarities, list))
for i in range(5):
    assert(isinstance(neymar_similarities[i], str))
```

**Task 14 (2 points).** Find a team with the highest win rate in the **2015/2016** season. Win rate means won matches / all matches. If there are several teams with the highest win rate return the first by name in lexical order

```python
matches_2015_1016 = match_data[(match_data['season'] == "2015/2016")]
```

```python
teams_2015_2016 = list(set(matches_2015_1016.loc[:,'home_team_api_id':'away_team_api_id'].to_numpy().flatten()))
teams_list_2015_2016 = [[x, 0, 0] for x in teams_2015_2016]

for i in range (len(matches_2015_1016)):
    home_team_api_id = matches_2015_1016.iloc[i]['home_team_api_id']
    away_team_api_id = matches_2015_1016.iloc[i]['away_team_api_id']

    home_win = 1 if matches_2015_1016.iloc[i]['home_team_goal'] > matches_2015_1016.iloc[i]['away_team_goal'] else 0
    away_win = 1 if matches_2015_1016.iloc[i]['home_team_goal'] < matches_2015_1016.iloc[i]['away_team_goal'] else 0

    teams_list_2015_2016[teams_2015_2016.index(home_team_api_id)][1] += 1
    teams_list_2015_2016[teams_2015_2016.index(home_team_api_id)][2] += home_win

    teams_list_2015_2016[teams_2015_2016.index(away_team_api_id)][1] += 1
    teams_list_2015_2016[teams_2015_2016.index(away_team_api_id)][2] += away_win
```

```python
team_win_rate = []

for i in range (len(teams_list_2015_2016)):
    team_win_rate.append([teams_list_2015_2016[i][0], teams_list_2015_2016[i][2]/teams_list_2015_2016[i][1]])

team_win_rate.sort(key=lambda x: x[1], reverse=True)
```

```python
team_win_name = team[team['team_api_id'] == team_win_rate[0][0]].iloc[0]['team_long_name']
tim_win_quantity = team_win_rate[0][1]
for i in range (1, len(team_win_rate)):
    team_name = team[team['team_api_id'] == team_win_rate[i][0]].iloc[0]['team_long_name']
    if team_win_rate[i][1] == tim_win_quantity:
        if team_name < team_win_name:
            team_win_name = team_name
    else:
        break
```

```python
team_highest_winrate_2015_2016 = team_win_name

print(team_highest_winrate_2015_2016)
```

**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
league_id = league_data[league_data['name'] == 'England Premier League'].iloc[0]['id']
match_2010_2011_england = match_data[(match_data['season'] == "2010/2011") & (match_data['league_id'] == league_id)]
```

```python
teams_2010_2011_england = list(set(match_2010_2011_england.loc[:,'home_team_api_id':'away_team_api_id'].to_numpy().flatten()))
teams_and_date_2010_2011_england = [[x, []] for x in teams_2010_2011_england ]

for i in range (len(match_2010_2011_england)):
    date = match_2010_2011_england.iloc[i]['date']
    home_team_index = teams_2010_2011_england.index(match_2010_2011_england.iloc[i]['home_team_api_id'])
    away_team_index = teams_2010_2011_england.index(match_2010_2011_england.iloc[i]['away_team_api_id'])
    teams_and_date_2010_2011_england[home_team_index][1].append(date)
    teams_and_date_2010_2011_england[away_team_index][1].append(date)

for i in range (len(teams_and_date_2010_2011_england)):
    teams_and_date_2010_2011_england[i][1] = [(datetime.strptime(x, '%Y-%m-%d %H:%M:%S')) for x in teams_and_date_2010_2011_england[i][1]]
    teams_and_date_2010_2011_england[i][1].sort()

for i in range (len(teams_and_date_2010_2011_england)):
    distance = []
    for j in range (len(teams_and_date_2010_2011_england[i][1]) - 1):
        distance.append((teams_and_date_2010_2011_england[i][1][j+1] - teams_and_date_2010_2011_england[i][1][j]).days)
    teams_and_date_2010_2011_england[i][1] = max(distance)

teams_and_date_2010_2011_england.sort(key=lambda x: x[1], reverse=True)
```

```python
highest_gap_england_2010_2011 = teams_and_date_2010_2011_england[0][1]

print(highest_gap_england_2010_2011)
```
